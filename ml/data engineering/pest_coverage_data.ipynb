```python
#intialize main dataFrame with country codes and names
countryCodeNameDict = {}
for key in countryBBs.keys():
    countryCodeNameDict[key] = countryBBs[key][0]
countryCodeNameDict
pestData = pd.DataFrame(data=countryCodeNameDict.items(), columns=["country code", "country name"])
pestData
```

```python
#Retrieves the pest risk level for each country. Each column is the risk level of a specific pest. Raw data, which will be analyzed later
id = "1"
north, south, west, east = "",  "", "", ""
stats = {}
for i in range(1,25): #24 layers accounts for the 24 keystone pests tracked by the API
    id = i
    stats = {}
    for country in countryBBs.keys():
        time.sleep(4) #sleeps between each api call to prevent limit throttling (1000/hr)
        coords = countryBBs[country][1] #retrieves coords of the bounding box
        west, south, east, north = coords[0], coords[1], coords[2], coords[3] #coords formatted [W, S, E, N]
        url = f'https://gems-exchange-pest.p.rapidapi.com/senay-eci/object/{id}/stats'
        coords = f'{west},{south},{east},{north}' 
        querystring = {"bbox":coords}
        stat = requestStats()
        if not stat or stat["mean"] == "NaN":
            stats[country] = 0
        else:
            stats[country] = stat["mean"]
    layerData = pd.DataFrame(stats.items(), columns=["country code", pestLayer[id]])
    pestData = pd.merge(pestData, layerData)
pestData
```

```python
#Retrieves the 5 pests with highest risk level of each country and gets the average to calculate the cumulative risk
pestAvg = {}
for i in range(0, pestData.shape[0]):
    row = pestData.loc[i]
    row.pop("country name")
    country = row.pop("country code")
    pestVals = row.values
    pestVals.sort()
    sum = 0
    for i in range(1, 3):
        sum += pestVals[len(pestVals) - i]
    avg = sum/5
    pestAvg[country] = avg
pestAverages = pd.DataFrame(pestAvg.items(), columns=["country code", "Cumulative risk"])
pestData = pestData.merge(pestAverages)
```
